```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import math

def NN_interpolation(img,dim): #dim is a 12 vector, representing the height and weight of interpolation
    scrH,scrW=img.shape
    retimg=np.zeros((dim[0],dim[1]),dtype=np.uint8)
    for i in range(dim[0]):
        for j in range(dim[1]):
            scrx=round(i*((scrH-1)/(dim[0]-1)))
            scry=round(j*((scrW-1)/(dim[1]-1)))
            retimg[i,j]=img[scrx,scry]
    return retimg
    
image=plt.imread('rice.tif')
plt.imshow(image,cmap=plt.cm.gray)
plt.show()
print(image.shape)
dim1=np.array([round(256*1.3),round(256*1.3)])
image1=NN_interpolation(image,dim1)
plt.imshow(image1,cmap=plt.cm.gray)
plt.show()
print(image1.shape)
savedimage1=Image.fromarray(image1)
savedimage1.save('enlarged_nearest_.tif')
dim2=np.array([round(256*(1-0.3)),round(256*(1-0.3))])
image2=NN_interpolation(image,dim2)
plt.imshow(image2,cmap=plt.cm.gray)
plt.show()
print(image2.shape)
savedimage2=Image.fromarray(image2)
savedimage2.save('shrunk_nearest_.tif')
```
