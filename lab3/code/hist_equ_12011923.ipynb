```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def hist_equ_12011923(input_image):
    H,W=input_image.shape
    total=H*W*1
    
    output_image=np.zeros([H,W],dtype=np.uint8)
    input_hist=[]
    output_hist=[]
    
    #input_histogram
    for i in range(256):
        input_hist.append(np.sum(input_image==i)/(total))
        
        
    sum_h=0
    for i in range(0,256):
        ind=np.where(input_image==i)
        sum_h+=len(input_image[ind])
        z_prime=round(255/total*sum_h)
        output_image[ind]=z_prime
        
        
    #output_histogram
    for i in range(256):
        output_hist.append(np.sum(output_image==i)/(total))
        
    return (output_image,input_hist,output_hist)

img=cv2.imread('Q3_1_2.tif',cv2.IMREAD_GRAYSCALE)
cv2.imshow('hhhhh',img)
cv2.waitKey(0)

print(img.shape)

x=np.arange(256)


(out,input_hist,output_hist)=hist_equ_12011923(img)
plt.imshow(out,cmap=plt.cm.gray)
plt.show()
savedimage1=Image.fromarray(out)
savedimage1.save('Q3_1_2_hist_equ_12011923.tif')
savedimage1.save('Q3_1_2_hist_equ_12011923.png')
print(out.shape)

#plt.subplot(2, 1, 1)
plt.bar(x,input_hist)
plt.title('histogram equalization of input image')
#plt.subplot(2, 1, 2)
#plt.bar(x,output_hist)
#plt.title('histogram equalization of output image')
#plt.close
```
